```python
import numpy as np
import math
```

```python
data = [
    [150, 7.0, 1, 'Apple'],
    [120, 6.5, 0, 'Banana'],
    [180, 7.5, 2, 'Orange'],
    [155, 7.2, 1, 'Apple'],
    [110, 6.0, 0, 'Banana'],
    [190, 7.8, 2, 'Orange'],
    [145, 7.1, 1, 'Apple'],
    [115, 6.3, 0, 'Banana']
]

encoding = {'Apple': 0 , 'Banana': 1 , 'Orange': 2}

weight = []
size = []
color = []
label = []

for i in data:
    weight.append(i[0])
    size.append(i[1])
    color.append(i[2])
    label.append(encoding.get(i[3]))
    

print(weight)
print(size)
print(color)
print(label)




```

```python
x = np.array([weight,size,color]).astype(float).T
y = np.array(label)

print(x)
print(y)
```

```python
def Eucledian(x,y):
    n = len(x)
    distance = 0
    for i in range(n):
        distance += (x[i]-y[i])**2

    return math.sqrt(distance)
```

```python
class KNNclassifier:
    def __init__(self,k=8):
        self.k = k 

    def fit(self,x,y):
        self.X_train = x
        self.Y_train = y

    def predict_one(self,x):
        distances = []
        for i in range(len(self.X_train)):
            dist = Eucledian(x,self.X_train[i])
            distances.append((dist, self.Y_train[i]))

        distances.sort(key = lambda item: item[0])
        k_nearest = distances[:self.k]

        k_labels = []

        for dist, labels in k_nearest:
            k_labels.append(labels)


        label_count = {}

        for i in k_labels:
            if i not in label_count:
                label_count[i] = 0
            label_count[i] += 1

        max_count = 0
        predicted_label = None



        for label, count in label_count.items():
            if count > max_count:
                max_count = count
                predicted_label = label

        return predicted_label
    
    def predict(self,X_test):
        predictions = []

        for x in X_test:
            pred = self.predict_one(x)
            predictions.append(pred)

        return np.array(predictions)

    
        

        

        
```

```python
test_data = np.array([
    [118, 6.2, 0],  # Expected: Banana(1)
    [160, 7.3, 1],  # Expected: Apple(0)
    [185, 7.7, 2]   # Expected: Orange(2)
])
test = KNNclassifier(k = 3)
test.fit(x,y)
predictions = test.predict(test_data)

print(f"Predictions: {predictions}")


```
